```python
import random # Para generar num. aleatorios en la A y B 
import math # Para operaciones matematicas como floor, ceil...
import multiprocessing as mp  # Para trabajar en paralelo
import time
import matplotlib.pyplot as pl # Para dibujar graficas
import os, platform as pt # Para obtener la informacion del dispositivo que ejecuta el programa
```

```python
def multiplicar_secuencial(matrizA, matrizB): # Funcion que calcula la multiplicacion de dos matrices secuencialmente (como lo haciamos desde 1º)
    matrizResultado = [[0] * num_columnas_B for i in range(num_filas_A)] # Generar dimensiones y llenar de 0s la matriz resultado
    for i in range(num_filas_A): # Hago la multiplicacion de AxB = C, i para iterar sobre las filas de A
        for j in range(num_columnas_B): # j para iterar sobre las columnas de B
            for k in range(num_columnas_A): # k para iterar en C
                matrizResultado[i][j] += matrizA[i][k] * matrizB[k][j] # Aqui se hace la multiplicación y la guardo en C.
    return matrizResultado

def multiplicar_paralelamente(matrizA, matrizB): # Funcion que prepara el reparto de trabajo para la mult. en paralelo
    num_cores = mp.cpu_count() # Obtengo los cores de mi pc
    size_col = math.ceil(num_columnas_B/num_cores) # Columnas a procesar x c/cpre, ver Excel adjunto
    size_fil = math.ceil(num_filas_A/num_cores) # Filas a procesar x c/cpre, ver Excel adjunto
    # Poner h para que sea short, 'i' para que sea int, 'b' para que sea binario...
    arrayCompartido = mp.RawArray('h', num_filas_A * num_columnas_B) # Array de memoria compartida donde se almacenaran los resultados, ver excel adjunto
    cores = [] # Array para guardar los cores y su trabajo
    for core in range(num_cores): # Asigno a cada core el trabajo que le toca, ver excel adjunto
        inicio_core = min(core * size_fil, num_filas_A) # Calculo el inicio del trabajo del core en relacion a las filas, ver excel
        final_core = min((core + 1) * size_fil, num_filas_A) # Calculo el fin del trabajo del core en relacion a las filas, ver excel
        cores.append(mp.Process(target=trabajo_de_un_core, args=(matrizA, matrizB, arrayCompartido, inicio_core, final_core))) # Añado al Array los cores y su trabajo
    for core in cores:
        core.start() # Arranco y ejecuto el trabajo para cada uno de los cores que tenga mi equipo, ver excel
    # for core in cores:
    #     core.join()# Bloqueo cualquier llamada hasta que terminen su trabajo todos los cores
    matrizCompartida = [[0] * num_columnas_B for i in range(num_filas_A)] # Convierto el array unidimensional MC en una matrix 2D (C_2D) 
    for i in range(num_filas_A):# i para iterar sobre las filas de A
        for j in range(num_columnas_B):# j para iterar sobre las columnas de B
            matrizCompartida[i][j] = arrayCompartido[i*num_columnas_B + j] # Guardo el C_2D los datos del array MC
    return matrizCompartida

def trabajo_de_un_core(matrizA, matrizB, arrayCompartido, inicio_core, final_core): # La tarea que hacen todos los cores
    for i in range(inicio_core, final_core): # Size representado en colores en el excel que itera sobre las filas en A
        for j in range(len(matrizB[0])): # Size representado en colores en el excel que itera sobre las columnas en B
            for k in range(len(matrizA[0])): # n_fil_B o lo que es l mismo el n_col_A
                arrayCompartido[i*len(B[0]) + j] += matrizA[i][k] * matrizB[k][j] # Guarda resultado en MC[] de cada core

if __name__ == '__main__':
    filasA = int(input("Filas de la matriz A:"))
    columnasA = int(input("Columnas de la matriz A:"))
    filasB = int(input("Filas de la matriz B:"))
    columnasB = int(input("Columnas de la matriz B:"))
    
    matrizA = [[random.randint(0,9) for i in range(columnasA)] for j in range(filasA)] # Genero matrizA con numeros aleatorios del 0 al 9, ver excel
    matrizB = [[random.randint(0,9) for i in range(columnasB)] for j in range(filasB)] # Genero matrizB con numeros aleatorios del 0 al 9, ver excel
    num_filas_A = len(matrizA) # Obtengo el numero de filas de matrizA 
    num_columnas_A = len(matrizA[0]) # Obtengo el numero de columnas de matrizA 
    num_filas_B = len(matrizB) # Obtengo el numero de filas de matrizB
    num_columnas_B = len(matrizB[0]) # # Obtengo el numero de columnas de matrizB
    if num_columnas_A != num_filas_B: raise Exception('Dimensiones no validas') # Compruebo que se puedan multiplicar A y B
    inicioSecuencial = time.time()
    multiplicar_secuencial(matrizA, matrizB) # Ejecuto multiplicacion secuencial
    finSecuencial = time.time()
    inicioParalelo = time.time()
    multiplicar_paralelamente(matrizA, matrizB) # Ejecuto multiplicacion paralela
    finParalelo = time.time()
    print('\nMatrices A y B multiplicadas en:\n-', finSecuencial-inicioSecuencial, 'segundos (secuencial)\n-', finParalelo-inicioParalelo, 'segundos (paralelo)\n') # Me funciona hasta 130, con 131 ya no responde
    
    nombres = ["Secuencial", "Paralelo"]
    datos = [finSecuencial-inicioSecuencial, finParalelo-inicioParalelo]
    
    grafica = pl.bar(nombres, datos)
    pl.title("Comparativa realizada en un PC " + str(pt.architecture()) + " con " + str(os.cpu_count()) + " cores")
    grafica
    pl.savefig('21938788_barras') # Guardar un png de la grafica con el nombre indicado en la ubicacion que indique "pwd"
```

### Estudio de efectividad en funcion de la complejidad de la tarea

```python
if __name__ == '__main__':
    
    dominio = 131
    intervalo = 2
    datos_secuenciales = [0]*int(dominio/intervalo)
    datos_paralelos = [0]*int(dominio/intervalo)
    tamanos = [0]*int(dominio/intervalo)
    it = 0
    
    for tamano in range(intervalo, dominio, intervalo):
        tamanos[it] = it*intervalo
        matrizA = [[random.randint(0,9) for i in range(tamano)] for j in range(tamano)] # Genero matrizA con numeros aleatorios del 0 al 9, ver excel
        matrizB = [[random.randint(0,9) for i in range(tamano)] for j in range(tamano)] # Genero matrizB con numeros aleatorios del 0 al 9, ver excel
        num_filas_A = len(matrizA) # Obtengo el numero de filas de matrizA 
        num_columnas_A = len(matrizA[0]) # Obtengo el numero de columnas de matrizA 
        num_filas_B = len(matrizB) # Obtengo el numero de filas de matrizB
        num_columnas_B = len(matrizB[0]) # # Obtengo el numero de columnas de matrizB
        if num_columnas_A != num_filas_B: raise Exception('Dimensiones no validas') # Compruebo que se puedan multiplicar A y B
        inicioS = time.time()
        multiplicar_secuencial(matrizA, matrizB) # Ejecuto multiplicacion secuencial
        finS = time.time()
        inicioP = time.time()
        multiplicar_paralelamente(matrizA, matrizB) # Ejecuto multiplicacion paralela
        finP = time.time()
        print("MATRICES " + str(tamano) + "x" + str(tamano) + ":\n- Secuencial: " + str(finS-inicioS) + "\n- En paralelo: " + str(finP-inicioP)) # Me funciona hasta 130

        nombres = ["Secuencial", "Paralelo"]
        datos_secuenciales[it] = finS-inicioS
        datos_paralelos[it] = finP-inicioP
        datos = [finS-inicioS, finP-inicioP]
        it+=1
        
grafica = pl.plot(tamanos, datos_secuenciales, label = "tiempos secuenciales")
grafica = pl.plot(tamanos, datos_paralelos, label = "tiempos paralelos")
pl.xlabel('x - tamaño de las matrices')
pl.ylabel('y - segundos necesitados')
pl.title("Rendimiento secuencial/paralelo segun la dificultad de la tarea")
pl.legend()
pl.savefig('21938788_lineal')
```
